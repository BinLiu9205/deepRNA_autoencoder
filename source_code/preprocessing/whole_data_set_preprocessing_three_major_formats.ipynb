```python
def transcriptsToGenes(df, indexJson):
    newDf = []
    df_trans = np.exp2(df)
    for key in indexJson.keys():
        indexNr = indexJson[key]
        df_tem = df_trans[:, indexNr]
        log_tem = df_tem.mean(axis=1)
        newDf.append(np.log2(log_tem))
    newDf = np.array(newDf)
    print(newDf.shape)
    newDf = np.array(newDf.transpose())
    return newDf
```

```python
gene_map = data['gene_to_index_relationship']
trimmed_gene_map = {}
reversed_gene_map = {}
for key in gene_map:
    if str(gene_map[key]) not in reversed_gene_map:
        trimmed_gene_map[key] = gene_map[key]
        reversed_gene_map[str(gene_map[key])] = key
```

```python
df_transcript = pd.read_pickle(
    '/mnt/dzl_bioinf/binliu/deepRNA/data_all_samples/all_samples_gene_level.pkl')
df_transcript_np = df_transcript.values
df_transcript_np = df_transcript_np.astype(float)
```

```python
res = np.absolute(np.corrcoef(df_transcript_np))
neigh = NearestNeighbors(n_neighbors=10, metric="precomputed")
neigh.fit(res)
rna_graph = neigh.kneighbors_graph(res)
rna_graph = rna_graph.toarray()
```

```python
# get the row, col indices of the non-zero elements in your adjacency matrix
conn_indices = np.where(rna_graph)

# get the weights corresponding to these indices
weights = rna_graph[conn_indices]

# a sequence of (i, j) tuples, each corresponding to an edge from i -> j
edges = zip(*conn_indices)

# initialize the graph from the edge sequence
G = ig.Graph(edges=edges, directed=True)
```

```python
partition = la.find_partition(
    G, la.CPMVertexPartition, resolution_parameter=0.02, seed=42)
```

```python
print(len(partition))
```

```python
reducted_gene_idx = []
rep_genes = []
for i in range(len(partition)):
    # if there is one gene in the community, then add the gene to the input
    if len(partition[i]) == 1:
        reducted_gene_idx.append(partition[i])
        rep_genes.append(df_transcript.index[partition[i]])
    else:
        # get all the elements in the community
        subgene = res[partition[i], :]
        subgene = subgene[:, partition[i]]
        # record the name of all the genes in the community
        sub_index = df_transcript.iloc[partition[i], :].index
        relationship_vec = np.sum(subgene, axis=0)

        max_idx = relationship_vec.tolist().index(max(relationship_vec.tolist()))
        reducted_gene_idx.append(partition[i][max_idx])
        rep_genes.append(sub_index[max_idx])
```
